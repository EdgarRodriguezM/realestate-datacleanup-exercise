#### Ejercicio 09. ¿Son los precios promedios de "Valdemorillo" y "Galapagar" los mismos? (★★☆)

Imprime ambos promedios y escribe una conclusión sobre ellos.

```python
valdemorillo = ds[ds['level5'] == 'Valdemorillo']
galapagar = ds[ds['level5'] == 'Galapagar']

promedio_valdemorillo = valdemorillo['price'].mean()
promedio_galapagar = galapagar['price'].mean()

print(f"Precio promedio en Valdemorillo: {promedio_valdemorillo:.2f} USD")
print(f"Precio promedio en Galapagar: {promedio_galapagar:.2f} USD")

if promedio_valdemorillo == promedio_galapagar:
    print("Los precios promedios son iguales.")
elif promedio_valdemorillo > promedio_galapagar:
    print("El precio promedio en Valdemorillo es mayor que en Galapagar.")
else:
    print("El precio promedio en Galapagar es mayor que en Valdemorillo.")
```

#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
import matplotlib.pyplot as plt

norm_df = cinturon_sur.copy()

norm_df['price_norm'] = norm_df.groupby('level5')['price'].transform(
    lambda x: (x - x.mean()) / x.std()
)

poblaciones_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

plt.figure(figsize=(10,6))
for poblacion in poblaciones_sur:
    plt.hist(
        norm_df[norm_df['level5'] == poblacion]['price_norm'],
        bins=20, alpha=0.5, label=poblacion
    )

plt.xlabel('Precio normalizado (Z-score)')
plt.ylabel('Frecuencia')
plt.title('Distribución de precios normalizados por población (Cinturón Sur)')
plt.legend()
plt.show()
```

```python
## Aquí: traza la coordenadas de los estados

## PON TU CÓDIGO AQUÍ:


from ipyleaflet import Map, CircleMarker, basemaps, LayerGroup
import random

# Poblaciones del cinturón sur
poblaciones_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]

# Filtrar el DataFrame
cinturon_sur = ds[ds['level5'].isin(poblaciones_sur)].copy()

# Convertir coordenadas a numérico (corrigiendo comas por puntos)
cinturon_sur['latitude'] = pd.to_numeric(cinturon_sur['latitude'].astype(str).str.replace(',', '.'), errors='coerce')
cinturon_sur['longitude'] = pd.to_numeric(cinturon_sur['longitude'].astype(str).str.replace(',', '.'), errors='coerce')

# Eliminar coordenadas nulas
cinturon_sur = cinturon_sur.dropna(subset=['latitude', 'longitude'])

# Crear mapa centrado en Madrid
mapa = Map(center=(40.35, -3.8), zoom=11.5, basemap=basemaps.OpenStreetMap.Mapnik)

# Colores por población
colores = {
    poblacion: "#{:06x}".format(random.randint(0, 0xFFFFFF))
    for poblacion in poblaciones_sur
}

# Grupo de marcadores
grupo_marcadores = LayerGroup()

# Añadir marcadores
for _, fila in cinturon_sur.iterrows():
    marcador = CircleMarker(
        location=(fila['latitude'], fila['longitude']),
        radius=6,
        color=colores.get(fila['level5'], "#000000"),
        fill_color=colores.get(fila['level5'], "#000000"),
        fill_opacity=0.6
    )
    grupo_marcadores.add_layer(marcador)

mapa.add_layer(grupo_marcadores)
mapa

```
